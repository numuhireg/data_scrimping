```python
import tweepy as tw
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import collections

import tweepy as tw
import nltk
from nltk.corpus import stopwords
import re
import networkx
from textblob import TextBlob

import warnings
warnings.filterwarnings("ignore")

##### On d√©finie une fonction de nettoyage des donn√©e

def remove_url(txt):
    """Remplace les url trouv√©e par le caract√®re vide " "
    

    Parametre
    ----------
    txt : string
        la variable string que l'on soutaite remplac√©

    Sortie
    -------
    
    Le fichier nettoy√© des url
    
    """

    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())


# On retire les caract√®res 
tweets_sans_urls = [remove_url(tweet.text)  for i in result for tweet in i]
```

```python
mots_dans_tweets = [tweet.lower().split() for tweet in tweets_sans_urls]
mots_dans_tweets
```

```python
# List of all words across tweets
ensemble_mots_pas_urls = list(itertools.chain(*mots_dans_tweets))

# Compte des mots
counts_no_urls = collections.Counter(ensemble_mots_pas_urls)

counts_no_urls.most_common(40)
```

```python
clean_tweets_no_urls = pd.DataFrame(counts_no_urls.most_common(15),
                             columns=['mots', 'compte'])

clean_tweets_no_urls.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

# Plot horizontal bar graph
clean_tweets_no_urls.sort_values(by='compte').plot.barh(x='mots',
                      y='compte',
                      ax=ax,
                      color="blue")

ax.set_title("Mots les plus pr√©sent dans les tweets")

plt.show()
```

```python
stop_words = set(stopwords.words('french'))

# View a few words from the set
list(stop_words)[0:25]
```

```python
#### Afficher le mots dans la liste de mots twwet si celui ci n'est pas dans stop word et ce pour chaque tweet


tweet_sans_sw = [[mots for mots in mots_tweet if not mots in stop_words]
                 for mots_tweet in mots_dans_tweets]
```

```python
tweet_sans_sw[0] #### R√©sultat du premier tweet sans stopword
```

```python
ens_mot_sans_sw = list(itertools.chain(*tweet_sans_sw))
compte_tw_no_sw = collections.Counter(ens_mot_sans_sw)
compte_tw_no_sw.most_common(15)
```

```python
tweet_filtre = pd.DataFrame(compte_tw_no_sw.most_common(30),
                             columns=['mots', 'compte'])

fig, ax = plt.subplots(figsize=(8, 8))

# Plot horizontal bar graph
tweet_filtre.sort_values(by='compte').plot.barh(x='mots',
                      y='compte',
                      ax=ax,
                      color="purple")

ax.set_title("Distibution des mots les plus pr√©sents (sans stops words) ")

plt.show()
```

```python
a_retirer = ["rt", "a", "plus","cette","fait","faire","sans","comment"]
#### On r√©utilise double listes
tweet_sans_mots_bannnis = [[mots for mots in mots_tweet if mots not in a_retirer]
                 for mots_tweet in tweet_sans_sw]
```

```python
tweet_filtre_2 = pd.DataFrame(compte_tw_no_smb.most_common(30),
                             columns=['mots', 'compte'])

fig, ax = plt.subplots(figsize=(8, 8))

tweet_filtre_2.sort_values(by='compte').plot.barh(x='mots',
                      y='compte',
                      ax=ax,
                      color="purple")

ax.set_title("Distibution des mots les plus pr√©sents (sans stops words et mots filtre) ")

plt.show()
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

from wordcloud import WordCloud, STOPWORDS
```

```python
# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Read the whole text.
text = open("gloriose.txt").read()# aha uyiha text ushakako igusomera attention il faut le telecharge en jupternotebook

# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open(path.join(d, "tete.jpg")))# aha uyiha image ushaka ko igusomera 

stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')

# generate word cloud
wc.generate(text)

# store to file
wc.to_file(path.join(d, "tete.jpg"))

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
#plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
#plt.axis("off")
plt.show()
```

```python
search_term = "Magnum"

tweets = tw.Cursor(api.search,
                   q=search_term,
                   lang="fr",
                   since='2020-01-07').items(1600)

# Remove URLs and create textblob object for each tweet
all_tweets_no_urls = [TextBlob(remove_url(tweet.text)) for tweet in tweets]

all_tweets_no_urls[:5]

# Calculate polarity of tweetsMagnum
magnum_sent_values = [[tweet.sentiment.polarity, str(tweet)] for tweet in all_tweets_no_urls]

# Create dataframe containing polarity values and tweet text
magnum_sent_df = pd.DataFrame(magnum_sent_values, columns=["polarity", "tweet"])
magnum_sent_df = magnum_sent_df[magnum_sent_df.polarity != 0]

magnum_sent_df.head(10)

fig, ax = plt.subplots(figsize=(8, 6))

magnum_sent_df.hist(bins=[-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1],
        ax=ax, color="purple")

plt.title("Sentiments from Tweets on Magnum")
plt.show()
```
